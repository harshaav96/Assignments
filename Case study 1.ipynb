```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.Read the dataset to the Python environment

```python
data=pd.read_csv(r"C:\Users\User\Desktop\cars_data.csv")
```

# 2.Check for the null values present in the dataset

```python
data.isna().sum()
```

# 3.Bar graph of male vs female buyers 

```python
plt.figure(figsize=(6,6))
sns.countplot(x='Buyer Gender',data=data)
plt.title('Male Buyers vs Female Buyers',fontsize=18)
plt.grid()
plt.show()
```

# 4.To find top 5 cars based on their sales price

```python
top5_cars=data.nlargest(5,['Sale Price'])
top5_cars
```

```python
top5_cars_df=pd.DataFrame(top5_cars,columns=['Make','Model','Country','Sale Price'])
top5_cars_df.iloc[:5]
```

# 5.To Find the least 5 cars based on their Resell price

```python
least5_cars=data.nsmallest(5,['Resell Price'])
least5_cars
```

```python
least5_cars_df=pd.DataFrame(least5_cars,columns=['Make','Model','Country','Resell Price'])
least5_cars_df.iloc[:5]
```
